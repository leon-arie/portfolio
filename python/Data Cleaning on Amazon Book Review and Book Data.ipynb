```python
import pandas as pd
import numpy as np
import re
```

### Load Amazon book reviews and ratings data. 

This data was taken directly from Amazon from 1996-2014. It contains 3 million unique book reviews.

We will start the project by first cleaning the data as there are a lot of random characters that we need to get rid of. This will insure that our model is more accurate as well as making it organized to load it into mySQL database. We also replace any null values and drop any columns we won't use.

This dataset contains title, profile name, review text, review summary, and ratings. We will use this data to create a recommender system for book suggestions.

```python
df=pd.read_csv("/Users/leona/OneDrive/Desktop/Books_rating.csv")
df = df.drop('Price', axis=1)
df=df.drop('review/time', axis=1)

df['User_id'] = df['User_id'].replace(np.nan, 'NA')
df['profileName'] = df['profileName'].replace(np.nan, 'NA')
df['review/text'] = df['review/text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))
df['review/summary'] = df['review/summary'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))
df['Title'] = df['Title'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))
df['profileName'] = df['profileName'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))

#df['review/summary'] = df['review/summary'].str.replace('=\\\\', 'NA', regex=True)
#df['review/summary'] = df['review/summary'].str.replace('\.{4}', '', regex=True)
#df['review/text'] = df['review/text'].str.replace('\\', '', regex=False)
#df['review/text'] = df['review/text'].str.replace('"', '')
#df['review/summary'] = df['review/summary'].str.replace('"', '')
#df['review/text'] = df['review/text'].str.replace('\.{4}', '', regex=True)
#df['profileName'] = df['profileName'].str.replace('\\', '')
```

```python
#Replace any Null Values in review/score column and default it to $0.00
for index, value in df['review/score'].items():
    if isinstance(value, str):
        df.at[index, 'review/score'] = '0.00'
```

### Next, load book data. 

This dataset was generated by Google API and contains information on 214K unique books. The data contains information about each book for example, title, author, image link, description and categories. We will use this data to join the book reviews table.

We will start by cleaning the data as there are a lot of random characters that we need to get rid of. This will insure that our model is more accurate as well as making it organized to load it into mySQL database. We also replace any null values and drop any columns we won't use.


```python
df2=pd.read_csv("/Users/leona/OneDrive/Desktop/books_data.csv")
```

```python
df2.head(25)
```

```python
df2['Title'] = df2['Title'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))
df2['description'] = df2['description'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))
df2['authors'] = df2['authors'].str.replace('[', '').str.replace(']', '').str.replace("'", '')
df2['categories'] = df2['categories'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', str(x)))


df2 = df2.drop('publishedDate', axis=1)
df2 = df2.drop('publisher', axis=1)
df2 = df2.drop('infoLink', axis=1)
df2 = df2.drop('ratingsCount', axis=1)
df2 = df2.drop('previewLink', axis=1)

df2['Title'] = df2['Title'].astype(str)
df2['description'] = df2['description'].astype(str)
df2['authors'] = df2['authors'].astype(str)
df2['image'] = df2['image'].astype(str)
df2['categories'] = df2['categories'].astype(str)

df2['Title'] = df2['Title'].replace('nan', 'NA')
df2['description'] = df2['description'].replace('nan', 'NA')
df2['authors'] = df2['authors'].replace('nan', 'NA')
df2['image'] = df2['image'].replace('nan', 'NA')
df2['categories'] = df2['categories'].replace('nan', 'NA')
```
